```python
import cv2
import numpy as np
import face_recognition
import os
```

```python
path='/content/gdrive/MyDrive/ImagesAttendance'
images=[]
classNames=[]
myList=os.listdir(path)
print(myList)
for cl in myList:
  curImg=cv2.imread(f'{path}/{cl}')
  images.append(curImg)
  classNames.append(os.path.splitext(cl)[0])
  
  #here we dont want to append name like bill gates.jpg we just want bill gates so we use os.path.spitext and split cl and take its 1st element
print(classNames)
#now well require a simple function that can convert all the encodings

def findEncodings(images):
  encodeList=[] #empty list that will contain all the encodings at the end
  for img in images:#to loop through all the images
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)#convert to rgb
    #now we need to find the encodings
    encode=face_recognition.face_encodings(img)[0]
    encodeList.append(encode)
  return encodeList

#to call the function
encodeListKnown= findEncodings(images)
print('Encoding Completed!')

#NOW WE NEED TO FIND MATCHES BETWEEN OUR ENCODINGS

cap=cv2.VideoCapture(0)
cv2.waitKey(1)
while True:
  
  success, img = cap.read()
  

  #now since we are doing this in real time to increase the speed we can reduce image size
  
  imgS=cv2.resize(img,(0,0),None,0.25,0.25)
  
  imgS=cv2.cvtColor(imgS,cv2.COLOR_BGR2RGB)
  #now to find encoding of our webcam
  facesCurFrame=face_recognition.face_locations(imgS)
  encodesCurFrame=face_recognition.face_encodings(imgS,facesCurFrame)

  #NOW WE FIND THE MATCHES
  #WE ITERATE THROUGH ALL THE FACES THAT WE FOUND IN CURRENTFRAME AND THEN COMPARE ALL OF THEM WITH ALL THE ENCODINGS

  for encodeFace, faceLoc in zip(encodesCurFrame, facesCurFrame):
    matches=face_recognition.compare_faces(encodeListKnown,encodeFace)
    faceDis=face_recognition.face_distance(encodeListKnown,encodeFace)
    print(faceDis)
    matchIndex=np.argmin(faceDis)

    if matches [matchIndex]:
      name=classNames[matchIndex].upper()
      print(name)


```
